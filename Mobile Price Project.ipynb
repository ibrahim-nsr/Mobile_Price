```python
x = df.drop('price_range', axis=1)
y = df['price_range']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=43)
```

```python
print('x_train shape: ', x_train.shape)
print('x_test shape: ', x_test.shape)
print('y_train shape: ', y_train.shape)
print('y_test shape: ', y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().fit(x_train)

x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)

x_train = np.array(x_train)
x_test = np.array(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
DT = DecisionTreeClassifier(criterion='entropy')
```

```python
DT.fit(x_train,y_train)
```

```python
prediction = DT.predict(x_test)
```

```python
DT.score(x_test,y_test)
```

```python
DT.score(x_train,y_train)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
DTcv = DecisionTreeClassifier()
params = {
    'criterion':['gini','entropy'],
    'max_depth':np.arange(1,11),
    'min_samples_split':np.arange(1,11),
    'min_samples_leaf':np.arange(1,11)
}
```

```python
grid_search = GridSearchCV(estimator=DTcv, param_grid=params, cv=5, return_train_score=True)
grid_search.fit(x_train, y_train)
```

```python
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

best_params
```

```python
y_pred = best_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, auc, roc_curve, classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
precision_score(y_test,y_pred, average='micro')
```

```python
recall_score(y_test,y_pred, average='micro')
```

```python
f1_score(y_test,y_pred, average='micro')
```
